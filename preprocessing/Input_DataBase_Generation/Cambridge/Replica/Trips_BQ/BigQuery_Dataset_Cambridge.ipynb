# All trips starting and ending in cambridge (BigQuery)

```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
filepath_Replica_BQ = ".././02 Data/Replica/Trips/BigQuery/Cambridge/"
replica_bq_df = pd.read_json(filepath_Replica_BQ + 'bq-results-20220714-203648-1657831026743.json', lines=True)
```

```python
drop_list=['person_id', 'household_id', 'mode','tour_type','timezone', 'transit_route_ids', 'network_link_ids']
replica_bq_df=replica_bq_df.drop(drop_list,axis=1)
```

```python
replica_bq_df['start_time']= (replica_bq_df['start_time']).astype(str)
replica_bq_df['end_time']= (replica_bq_df['end_time']).astype(str)
replica_bq_df['start_time']= replica_bq_df['start_time'].str[11:19]
replica_bq_df['end_time']= replica_bq_df['end_time'].str[11:19]
replica_bq_df.head()
```
